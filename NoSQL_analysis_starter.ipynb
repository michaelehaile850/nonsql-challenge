### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score


penang_flavours = establishments.find_one({"BusinessName": "Penang Flavours"})

degree_search = 0.01
latitude = penang_flavours["geocode"]["latitude"] if penang_flavours else None
longitude = penang_flavours["geocode"]["longitude"] if penang_flavours else None

if latitude is not None and longitude is not None:
    query = {
        "RatingValue": 5,
        "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
        "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
    }

sort = [("scores.Hygiene", 1)]
limit = 5
results = establishments.find(query).sort(sort).limit(limit)
```

```python
# Convert result to Pandas DataFrame
results = establishments.find(query).sort(sort).limit(limit)
df=pd.DataFrame(list(results))
df.head(1)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create 

# Aggregation pipeline to count establishments with a hygiene score of 0 in each local authority
pipeline = [
    {
        '$match': {
            'scores.Hygiene': 0  # Match establishments with a hygiene score of 0
        }
    },
    {
        '$group': {
            '_id': '$LocalAuthorityName',  # Group by local authority name
            'count': {'$sum': 1}  # Count the number of establishments in each group
        }
    },
    {
        '$sort': {'count': -1}   Sort the results by count in descending order
    }
]

# Execute the aggregation pipeline
results = list(establishments.aggregate(pipeline))

# Print the results
for result in results:
    print(f"Local Authority: {result['_id']}, Establishments with Hygiene Score of 0: {result['count']}")


```

```python
# Convert the result to a Pandas DataFrame

results = list(establishments.aggregate(pipeline))
df = pd.DataFrame(results)

# Rename columns for clarity
df.rename(columns={'_id': 'LocalAuthorityName', 'count': 'NumberOfEstablishments'}, inplace=True)

#Number of row
num =len(df)
num

# Display the DataFrame
df.head(10)
```
